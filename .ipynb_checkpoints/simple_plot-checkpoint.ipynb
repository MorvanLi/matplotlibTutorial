3. easy to plot 2D, 3D data;

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)

x = np.linspace(-1, 1, 50)
y = 2*x + 1
plt.plot(x, y)
plt.show()
```
